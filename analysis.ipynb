```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data\\life_expectancy_data.csv")
print(df.head(10))
```

```python
drop = df.drop_duplicates(subset=["Country"])
print(drop.head(15))
# normalize column names then plot Year vs Life expectancy
drop = drop.copy()
drop.columns = drop.columns.str.strip()
drop.plot(x="Year", y="Life expectancy", kind="bar")
plt.xlabel("Year")
plt.ylabel("Life expectancy")
plt.title("Life Expectancy vs Year")
plt.show()
```
