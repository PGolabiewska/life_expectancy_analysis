```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data\\life_expectancy_data.csv")
print(df.head(10))
```

```python
poland_alcohol = df[(df["Country"] == "Poland") &df["Alcohol"].notna()][["Year", "Alcohol"]].sort_values("Year")

plt.figure(figsize=(10, 5))
plt.bar(poland_alcohol["Year"], poland_alcohol["Alcohol"], color="skyblue")

plt.title("Alcohol consumption in Poland by year")
plt.xlabel("Year")
plt.ylabel("Alcohol")
plt.show()
```

```python
country_data = df[df["Year"] == 2015].copy()
country_data = country_data.dropna(subset=["Life expectancy "])

best = country_data.sort_values("Life expectancy ", ascending=False).head(10)
worst = country_data.sort_values("Life expectancy ", ascending=True).head(10)

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

axes[0].bar(best["Country"], best["Life expectancy "], color="#4C78A8")
axes[0].set_title("Top 10 countries (2015)")
axes[0].set_ylabel("Life expectancy")
axes[0].tick_params(axis="x", rotation=45)
for label in axes[0].get_xticklabels():
    label.set_ha("right")

axes[1].bar(worst["Country"], worst["Life expectancy "], color="#E45756")
axes[1].set_title("Worst 10 countries (2015)")
axes[1].set_ylabel("Life expectancy")
axes[1].tick_params(axis="x", rotation=45)
for label in axes[1].get_xticklabels():
    label.set_ha("right")

plt.tight_layout()
plt.show()
```
